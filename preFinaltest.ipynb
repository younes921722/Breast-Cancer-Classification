###                      Data spliting 

```python
X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)
```

```python
# scaling data
from sklearn.preprocessing import StandardScaler
s = StandardScaler()

X_train = s.fit_transform(X_train)
X_test  = s.fit_transform(X_test)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression

logModel = LogisticRegression()
logModel.fit(X_train, y_train)
pred = logModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Confusion matrix without normalization
print("Confusion matrix")

# Make predictions on the test set
y_pred = logModel.predict(X_test)

# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Display the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=logModel.classes_)
disp.plot(cmap='Blues')
plt.show()  # Make sure to import matplotlib.pyplot as plt

# Create a normalized confusion matrix
Norm_conf_matrix = confusion_matrix(y_test, y_pred, normalize='true')

# Display the normalized confusion matrix 
disp = ConfusionMatrixDisplay(confusion_matrix=Norm_conf_matrix, display_labels=logModel.classes_)
disp.plot(cmap='Blues')
plt.show()
```

```python
print(conf_matrix)
print(classification_report(y_test, y_pred))
logisticReg_accuracy = accuracy_score(y_test, y_pred)
print(logisticReg_accuracy)
```

# Neural Network

```python
# importing tensorflow and Keras
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

NN_model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(60, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

NN_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network

history = NN_model.fit(X_train, y_train, validation_split=0.1, epochs=20)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, NN_accuracy = NN_model.evaluate(X_test, y_test)
print(NN_accuracy)
```

```python
# converting the prediction probability to class labels
# Make predictions on the test set
y_pred = NN_model.predict(X_test)
Y_pred_labels = [np.argmax(i) for i in y_pred]
Y_pred_labels = [np.argmax(i) for i in y_pred]
print(Y_pred_labels)
```

```python
print("Confusion matrix")

# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, Y_pred_labels)

# Display the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels= [0,1])
disp.plot(cmap='Greens')
plt.show()
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

gbc = GradientBoostingClassifier()

parameters = {
    'loss': ['deviance', 'exponential'],
    'learning_rate': [0.001, 0.1, 1, 10],
    'n_estimators': [100, 150, 180, 200]
}

grid_search_gbc = GridSearchCV(gbc, parameters, cv = 5, n_jobs = -1, verbose = 1)
grid_search_gbc.fit(X_train, y_train)
```

```python
# best parameters 
grid_search_gbc.best_params_
```

```python
# best score
grid_search_gbc.best_score_
```

```python
gbc = GradientBoostingClassifier(learning_rate = 0.1, loss = 'exponential', n_estimators = 180)
gbc.fit(X_train, y_train)
```

```python
y_pred = gbc.predict(X_test)
# accuracy score
print(accuracy_score(y_train, gbc.predict(X_train)))

gbc_acc = accuracy_score(y_test, y_pred)
print(gbc_acc)
```

```python
print("Confusion matrix")

# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Display the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels= [0,1])
disp.plot(cmap='Greens')
plt.show()
```
